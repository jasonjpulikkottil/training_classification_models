Task 1

```python

import pandas as pd

# read csv file
df = pd.read_csv("patients.csv")

# display DataFrame
print(df.head())


```

Task 2

```python

cols = ['Age','Gender','Systolic','Diastolic','Smoker']

data = df[cols].copy()

print(data.head(5))

X = data.loc[:, data.columns.drop(['Smoker'])]
Y = data.Smoker
```

Task 3

```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn import preprocessing


le = preprocessing.LabelEncoder()
le.fit(X.Gender)
X.Gender=le.transform(X.Gender)



# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1) # 70% training and 30% test



# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

```

Task 4

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.scatter(data['Systolic'],data['Diastolic'],c=Y,cmap='Spectral')
plt.colorbar()
plt.title('Scatter plot')
plt.xlabel('X - Systolic')
plt.ylabel('Y - Diastolic')
plt.show()


plt.scatter(data['Gender'],data['Diastolic'],c=Y,cmap='Spectral')
plt.colorbar()
plt.title('Scatter plot')
plt.xlabel('X - Gender')
plt.ylabel('Y - Diastolic')
plt.show()


plt.scatter(data['Age'],data['Diastolic'],c=Y,cmap='Spectral')
plt.colorbar()
plt.title('Scatter plot')
plt.xlabel('X - Age')
plt.ylabel('Y - Diastolic')
plt.show()




# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

Accuracy is 0.8333333333333334 which is good.
gender doesn't have a better clustering, but age have a better clustering.
